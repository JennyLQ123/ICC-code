```python
import matplotlib.pyplot 
import pylab as plt
import matplotlib
import numpy as np
x=[]
y=[]
with open("FB_CDF.txt",'r') as t:
    l=[t.strip().split() for t in t.readlines()]
l[0][0]=1
for node in l:
    x.append(int(node[0]))
    y.append(float(node[1]))

font = {'size':16}
matplotlib.rc('font',**font)
plt.figure(figsize=(6, 4))
plt.subplot(1,1,1)
plt.semilogx(x,y,color="crimson",linewidth=2,linestyle='-.')
plt.grid(True)

plt.xlabel("Flow Size(Bytes)")
plt.ylabel("CDF")
plt.tight_layout()
plt.savefig('cdf.pdf', format='pdf' ,dpi=1000)
plt.show()
```
