```python
import matplotlib.pyplot as plt
import networkx as nx

RG=nx.random_graphs.random_regular_graph(3,20)
pos=nx.spectral_layout(RG)
nx.draw(RG,pos,with_labels=False,node_size=50)
print(len(RG.edges()))
plt.show()
```

```python
#topo.pdf
import matplotlib.pyplot as plt
import networkx as nx
G=nx.Graph()
for i in range(20):
    G.add_node(i)
for i in range(19):
    if i%4==3:
        pass
    else:
        G.add_edge(i,i+1)
for i in range(16):
    G.add_edge(i,i+4)
pos=nx.random_layout(G)
nx.draw(G,pos,with_labels=False,node_size=100,width=1)
plt.savefig('pic/topo.pdf', format='pdf' ,dpi=1000,bbox_inches="tight")
#plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
ER=nx.random_graphs.erdos_renyi_graph(20,0.2)
pos=nx.random_layout(ER)
nx.draw(ER,pos,with_labels=False,node_size=50)
print(ER.number_of_edges())
plt.show()
```

```python
WS = nx.random_graphs.watts_strogatz_graph(20, 3, 0.3)
pos = nx.spectral_layout(WS)
nx.draw(WS, pos, with_labels=False, node_size=30)
print(WS.number_of_edges())
plt.show()
```

```python
BA = nx.random_graphs.barabasi_albert_graph(20, 1)
pos = nx.spring_layout(BA)
nx.draw(BA, pos, with_labels=False, node_size=30, node_color='black')
print(BA.number_of_edges())
plt.show()
```
